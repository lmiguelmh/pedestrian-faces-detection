```python
# show some images
import os 
path = 'D:\\projects\\tesis\\pedestrian-faces-detection\\datasets\\INRIAPerson\\96X160H96\\Train\\pos\\'
inria_list = os.listdir(path)
print(len(inria_list), " train images")
```

```python
% matplotlib inline

from matplotlib import pyplot as plt
import cv2

img_path = path + inria_list[0]
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_height = img.shape[0]
img_width = img.shape[1]

print('PATH\t:', img_path)
print('SIZE\t: %dx%d' % (img_width, img_height))

plt.imshow(img, interpolation='none')
plt.xticks([]), plt.yticks([])
plt.show()
```
